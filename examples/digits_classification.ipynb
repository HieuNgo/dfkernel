```python
# Standard scientific Python imports
import matplotlib.pyplot as plt

# Import datasets, classifiers and performance metrics
from sklearn import datasets, svm, metrics

# The digits dataset
datasets.load_digits()
```

```python
# The data that we are interested in is made of 8x8 images of digits, let's
# have a look at the first 4 images, stored in the `images` attribute of the
# dataset.  If we were working from image files, we could load them using
# matplotlib.pyplot.imread.  Note that each image must have the same size. For these
# images, we know which digit they represent: it is given in the 'target' of
# the dataset.
list(zip(Out['a51669'].images, Out['a51669'].target))
```

```python
for index, (image, label) in enumerate(Out['b964b9'][:4]):
    plt.subplot(2, 4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Training: %i' % label)

# To apply a classifier on this data, we need to flatten the image, to
# turn the data in a (samples, feature) matrix:
len(Out['a51669'].images)
```

```python
Out['a51669'].images.reshape((Out['64cf96'], -1))
```

```python
# Create a classifier: a support vector classifier
svm.SVC(gamma=0.001)
```

```python
# We learn the digits on the first half of the digits
Out['9e89cf'].fit(Out['7a1c6c'][:Out['64cf96'] // 2], Out['a51669'].target[:Out['64cf96'] // 2])
```

```python
# Now predict the value of the digit on the second half:
expected = Out['a51669'].target[Out['64cf96'] // 2:]
predicted = Out['9e89cf'].predict(Out['7a1c6c'][Out['64cf96'] // 2:])
```

```python
print("Classification report for classifier %s:\n%s\n"
      % (Out['9e89cf'], metrics.classification_report(expected, predicted)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(expected, predicted))
list(zip(Out['a51669'].images[Out['64cf96'] // 2:], predicted))
```
